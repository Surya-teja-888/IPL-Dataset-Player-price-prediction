```python
# Importing Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")
```

```python
image = cv2.imread("/kaggle/input/traffic-sign-cropped/crop_dataset/crop_dataset/00013/00014_00019.jpg")
plt.imshow(image)
```

```python
train_path = "../input/traffic-sign-cropped/crop_dataset/crop_dataset"
test_path = "../input/traffic-sign-cropped/test_data"
```

```python
import keras
from keras import layers
from keras.models import Sequential
from keras.preprocessing.image import ImageDataGenerator
```

# Data Preprocessing

```python
train_datagen = ImageDataGenerator(rescale = 1./225,validation_split = 0.2,horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale  =1./255)
```

```python
train_images = train_datagen.flow_from_directory(train_path,target_size = (224,224),class_mode = "categorical",subset = "training")
val_images = train_datagen.flow_from_directory(train_path,target_size = (224,224),class_mode = "categorical",subset = "validation")
test_images = test_datagen.flow_from_directory(test_path,target_size = (224,224),class_mode = "categorical")
```

# Building Neural Network

```python
# Building Neural Network
model = keras.Sequential([
    keras.layers.Conv2D(input_shape = (224,224,3),filters = 32,kernel_size = (3,3),padding = "same",activation = "relu"),
    keras.layers.MaxPool2D(pool_size = (2,2),strides = (2,2)),
    keras.layers.Conv2D(filters = 64,kernel_size = (3,3),activation = "relu",padding = "same"),
    keras.layers.MaxPool2D(pool_size = (2,2),strides = (2,2)),
    keras.layers.GlobalAveragePooling2D(),
    keras.layers.Dense(128,activation = "relu"),
    keras.layers.Dense(128,activation = "relu"),
    keras.layers.Dense(43,activation = "softmax")
])
```

```python
model.summary()
```

```python
# Compiling the Model
model.compile(optimizer = "adam",loss = "categorical_crossentropy",metrics = ["accuracy"])
```

```python
# Fitting the model
history = model.fit(train_images,validation_data = val_images,epochs = 10)
```

```python
y_pred = model.predict(test_images)
```

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Accuracy Graphs")
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.legend(["accuracy","val_accuracy"])
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Loss Graphs")
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.legend(["loss","val_loss"])
```
