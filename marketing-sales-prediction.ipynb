# Importing Modules

```python
# Importing Modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

# Loading the Data

```python
# Loding the data
data = pd.read_csv("/kaggle/input/dummy-advertising-and-sales-data/Dummy Data HSS.csv")
data
```

```python
# Getting info of the data
data.info()
```

# Preprocessing the Data

```python
# Checking Null values
data.isnull().sum()
```

```python
# target column contains Null values
# We drop the rows of target columns which contain Null values
missing_values = data[data["Sales"].isna()].index
missing_values
```

```python
data_1 = data.drop(missing_values,axis = 0)
data_1.isnull().sum()
```

```python
# Now Fill the NaN values in the "TV","Radio","Social Media", columns with mean()
columns = data_1[["TV","Radio","Social Media"]]
for columns in columns:
    data_1[columns].fillna(data_1[columns].mean(),inplace = True) # Inplace = True gives the modified data
```

```python
data_1.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import metrics
from sklearn import preprocessing
```

```python
data_1["Influencer"].value_counts()
```

```python
le = preprocessing.LabelEncoder()
data_1["Influencer"] = le.fit_transform(data_1["Influencer"])
```

# Building Model

```python
# Now Building Model
x = data_1.drop("Sales",axis = 1)
y = data_1["Sales"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
```

```python
lg = LinearRegression()
model = lg.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
# Building Model using gradient boosting
gd = GradientBoostingRegressor()
model_1 = gd.fit(x_train,y_train)
```

```python
y_pred_1 = model_1.predict(x_test)
y_pred_1
```

```python
metrics.mean_squared_error(y_test,y_pred_1)
```
