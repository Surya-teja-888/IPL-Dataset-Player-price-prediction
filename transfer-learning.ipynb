```python
# Importing Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")
```

```python
image = cv2.imread("/kaggle/input/natural-images/natural_images/motorbike/motorbike_0086.jpg")
plt.imshow(image)
```

```python
image_dir = Path("../input/natural-images")
image_dir
```

# Data Preprocessing

```python
files = list(image_dir.glob(r"**/*.jpg"))
files
```

```python
# Getting labels from files
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1],files))
labels
```

```python
# Making DataFrame
image_data = pd.DataFrame({"FilePath":files,"Labels":labels}).astype(str)
image_data
```

```python
import tensorflow as tf 
from sklearn.model_selection import train_test_split
```

```python
train_data,test_data = train_test_split(image_data,train_size = 0.7,random_state = 42)
```

```python
train_data
```

```python
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1./255,horizontal_flip = True,validation_split = 0.2)
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1./255)
```

```python
train_images = train_datagen.flow_from_dataframe(dataframe = train_data,target_size = (224,224),x_col = "FilePath",y_col = "Labels",class_mode = "categorical",subset = "training")
val_images = train_datagen.flow_from_dataframe(dataframe = train_data,traget_size = (224,224),x_col = "FilePath",y_col = "Labels",class_mode = "categorical",subset = "validation")
test_images = test_datagen.flow_from_dataframe(dataframe = test_data,targte_size = (224,224),x_col = "FilePath",y_col = "Labels",class_mode = "categorical")
```

# Building the Model

```python
inception = tf.keras.applications.InceptionV3(include_top = False,weights = "imagenet",input_shape = (224,224,3))
for layers in inception.layers:
    layers.trainable = False
```

```python
x = tf.keras.layers.GlobalAveragePooling2D()(inception.output)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(528,activation = "relu")(x)
x = tf.keras.layers.Dense(528,activation = "relu")(x)
outputs = tf.keras.layers.Dense(8,activation = "softmax")(x)
model = tf.keras.models.Model(inputs = inception.input,outputs = outputs)
```

```python
# Compiling the Model
model.compile(optimizer = "adam",loss = "categorical_crossentropy",metrics = "accuracy")
```

```python
# Fitting the Model
history = model.fit_generator(train_images,validation_data = val_images,epochs = 5)
```

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Accuracy_Graphs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Accuracy","Val_Accuracy"])
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Loss_Graphs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(["loss","Val_loss"])
```
