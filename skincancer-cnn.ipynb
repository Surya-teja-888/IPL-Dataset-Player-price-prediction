```python
from sklearn.utils import resample
```

```python
# Upsampling  
# Since the data is imbalanced we are doing upsampling
label_0 = data[data["label"] == 0]
label_1 = data[data["label"] == 1]
label_2 = data[data["label"] == 2]
label_3 = data[data["label"] == 3]
label_4 = data[data["label"] == 4]
label_5 = data[data["label"] == 5]
label_6 = data[data["label"] == 6]
label_0_upsampled = resample(label_0,replace = True,n_samples = 6690)
label_1_upsampled = resample(label_1,replace = True,n_samples = 6690)
label_2_upsampled = resample(label_2,replace = True,n_samples = 6690)
label_3_upsampled = resample(label_3,replace = True,n_samples = 6690)
label_4_upsampled = resample(label_4,replace = True,n_samples = 6690)
label_5_upsampled = resample(label_5,replace = True,n_samples = 6690)
label_6_upsampled = resample(label_6,replace = True,n_samples = 6690)
new_data = pd.concat([label_0_upsampled,label_1_upsampled,label_2_upsampled,label_3_upsampled,label_4_upsampled,label_5_upsampled,label_6_upsampled])
new_data.head()
```

```python
image_dir = Path("../input/skin-cancer-mnist-ham10000")
files = list(image_dir.glob(r"**/*.jpg"))
files
```

```python
# Importing Modules needed for building neural network
import keras
from keras.models import Sequential, Model
from keras import layers
from keras.preprocessing.image import ImageDataGenerator
```

```python
# Data Agumentation
train_datagen = ImageDataGenerator(rescale = 1./255,validation_split = 0.2)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_data = train_datagen.flow_from_directory(image_dir,
                                               target_size = (224,224),
                                               batch_size = 64,
                                               class_mode = "categorical",
                                               subset = "training"
)
validation_data = train_datagen.flow_from_directory(image_dir,
                                                    target_size = (224,224),
                                                    batch_size = 64,
                                                    class_mode = "categorical",
                                                    subset = "validation"
)
test_data = test_datagen.flow_from_directory(image_dir,
                                             target_size = (224,224),
                                             batch_size = 64,
                                             class_mode = "categorical"
)
```

```python
# Bulding the Model
model = keras.Sequential([
                        keras.layers.Conv2D(input_shape = (224,224,3),filters = 64,kernel_size = (3,3),padding = "same",activation = "relu"),
                        keras.layers.Conv2D(filters = 64,kernel_size = (3,3),padding = "same",activation = "relu"),
                        keras.layers.MaxPool2D(pool_size = (2,2)),
                        keras.layers.Conv2D(filters = 128,kernel_size = (3,3),padding = "same",activation = "relu"),
                        keras.layers.MaxPool2D(pool_size = (2,2),strides = (2,2)),
                        keras.layers.Conv2D(filters = 256,kernel_size = (3,3),padding = "same",activation = "relu"),
                        keras.layers.MaxPool2D(pool_size = (2,2),strides = (2,2)),
                        keras.layers.GlobalAveragePooling2D(),
                        keras.layers.Dense(128,activation = "relu"),
                        keras.layers.Dense(128,activation = "relu"),
                        keras.layers.Dense(4,activation = "softmax")
])
```

```python
# Compiling the Model
model.compile(optimizer = "adam",loss = "categorical_crossentropy",metrics = ["accuracy"])
```

```python
# Fitting the model
history = model.fit_generator(train_data,validation_data = validation_data,epochs = 5,steps_per_epoch = len(train_data),validation_steps = len(validation_data))
```

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Accuracy_Graphs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Accuracy","Val_Accuracy"])
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Loss_Graphs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(["loss","Val_loss"])
```
