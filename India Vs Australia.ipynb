```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("vaccination_progress.csv")
data.head()
```

```python
# We analyse vaccination in India and Australia
data_ind = data[data["country"] == "India"]
data_aus = data[data["country"] == "Australia"]
```

```python
numerical_columns = (data_ind.dtypes == "float")
numerical_list = list(numerical_columns[numerical_columns].index)
print(numerical_list)
```

```python
# Filling the null values
for columns in numerical_list:
    data_ind[columns].fillna(data_ind[columns].mean(),inplace = True)
```

```python
numerical_columns = (data_aus.dtypes == "float")
numerical_list = list(numerical_columns[numerical_columns].index)
print(numerical_list)
```

```python
# Filling the null values
for columns in numerical_list:
    data_aus[columns].fillna(data_aus[columns].mean(),inplace = True)
```

```python
import datetime
data_ind["date"] = pd.to_datetime(data_ind["date"])
data_aus["date"] = pd.to_datetime(data_aus["date"])
```
